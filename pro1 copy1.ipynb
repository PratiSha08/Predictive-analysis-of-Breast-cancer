```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve
```

# Decision tree (model1)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
model1=DecisionTreeClassifier()
```

```python
model1.fit(xtrain,ytrain)
plt.figure(figsize=(35,15))
plot_tree(model1,feature_names=xtrain.columns,class_names=['Malignant','Benign'],filled=True,) #Plotting the tree
plt.show()
```

```python
print("Decision Tree accuracy (on training dataset)= ",model1.score(xtrain,ytrain))
```

```python
print(classification_report(ytest,model1.predict(xtest)))
print("Accuracy Score= ",accuracy_score(ytest,model1.predict(xtest)))
print("Confusion Matrix:\n",confusion_matrix(ytest,model1.predict(xtest)))
```

```python
y_pred = model1.predict(xtest)

y_pred = np.array(y_pred>0.5)
print(y_pred.ravel().astype(int))
```

```python
np.array(ytest)
```

```python
def plot_roc_curve(model,xtest,ytest):
    ypred=model.predict(xtest)
    fpr,tpr,thresholds=roc_curve(ytest,ypred)
    plt.plot([0,1],[0,1],color="green",linestyle='--')
    plt.plot(fpr,tpr)
    plt.xlabel("False positive rate",color='green')
    plt.ylabel("True positive rate",color='green')
    plt.title("ROC curve")
```

```python
plot_roc_curve(model1,xtest,ytest)
```

## NEURAL NETWORK MODEL 

```python
# pip install tensorflow
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras import optimizers
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#Creating ANN model
ann=Sequential()
```

building layers

```python
#Creating input and first hiddenn layer
ann.add(Dense(input_shape=(xtest.shape[1],),units=20,activation='relu',kernel_initializer='uniform'))
```

```python
#creating second hidden layer
ann.add(Dense(units=15,activation='relu',kernel_initializer='uniform'))
```

```python
#creaating output layer
ann.add(Dense(units=1,activation='sigmoid',kernel_initializer='uniform'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Stopping the training if there's no improvement in the accuracy
es=EarlyStopping(monitor='accuracy',mode='max',patience=50,restore_best_weights=True)
```

```python
model=ann.fit(xtrain,ytrain,batch_size=100,epochs=500,callbacks=[es])
```

```python
# summary of all data in history
plt.plot(model.history['accuracy'])
plt.title('history of model accuracy curve\n',color='blue')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training'], loc='best')
plt.show()
```

```python
y_pred = ann.predict(xtest)

y_pred = np.array(y_pred>0.5)
print(y_pred.ravel().astype(int))
```
